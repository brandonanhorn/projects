```python
import pandas as pd
```

```python
df_intake = pd.read_csv("data/Austin_Animal_Center_Intakes.csv")
df_outcomes = pd.read_csv("data/Austin_Animal_Center_Outcomes.csv")
```

```python
df_intake.columns = [i.lower() for i in df_intake.columns]
df_outcomes.columns = [i.lower() for i in df_outcomes.columns]
```

```python
df_dog_intake = df_intake[(df_intake["animal type"] == "Dog")]
df_dog_outcome = df_outcomes[(df_outcomes["animal type"] == "Dog")]
```

```python
df_cat_intake = df_intake[(df_intake["animal type"] == "Cat")]
df_cat_outcome = df_outcomes[(df_outcomes["animal type"] == "Cat")]
```

```python
len(df_dog_intake), len(df_dog_outcome)
```

```python
df_dog_outcome.head(1)
```

```python
df_dogs = pd.merge(df_dog_intake, df_dog_outcome, how="left", on="animal id")
```

```python
df_cats = pd.merge(df_cat_intake, df_cat_outcome, how="inner", on="animal id")
```
