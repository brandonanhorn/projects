```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso
from sklearn_pandas import DataFrameMapper
from sklearn.metrics import mean_squared_error

pd.set_option("max.columns", None)
```

```python
df = pd.read_csv("data/CONVENIENT_global_confirmed_cases.csv")
```

```python
df.tail()
```

```python
df_2 = df[["Country/Region", "US"]]
```

```python
df_2 = df_2.rename(columns={"Country/Region": "date", "US":"total"})
```

```python
df_2
```

```python
df_2 = df_2.iloc[1:]
```

```python
df_2['date'] = pd.to_datetime(df_2['date'])
```

```python
df_2.dtypes
```

```python
df_2["previous_day_total"] = df_2["total"].shift()
```

```python
df_2 = df_2.fillna(0)
```

```python
target = "total"
X = df_2.drop(target, axis=1)
y = df_2[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
mapper = DataFrameMapper([
    (["previous_day_total"], StandardScaler())
],df_out=True)
```

```python
Z_train = mapper.fit_transform(X_train)
```

```python
Z_test = mapper.transform(X_test)
```

```python
model = LinearRegression().fit(Z_train,y_train)
print("LinearRegression train score is " + str(model.score(Z_train,y_train)))
print("LinearRegression test score is " + str(model.score(Z_test,y_test)))

print("Mean squared error is " + str(mean_squared_error(y_test, model.predict(Z_test))**(1/2)))
```

```python
model = Lasso().fit(Z_train,y_train)
print("Lasso train score is " + str(model.score(Z_train,y_train)))
print("Lasso test score is " + str(model.score(Z_test,y_test)))

print("Mean squared error is " + str(mean_squared_error(y_test, model.predict(Z_test))**(1/2)))
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(mapper, model)
```

```python
pipe.fit(X_train, y_train)
```
