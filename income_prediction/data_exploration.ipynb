```python
df = pd.read_csv("data/application_record.csv")
```

```python
df.columns = [i.lower() for i in df.columns]
```

```python
df.head(1)
```

```python
df.info()
```

```python
age = []
for i in df["days_birth"]:
    z = i / 365
    age.append(z)
```

```python
df["age"] = age
```

```python
df["age"] = df["age"].abs().astype("int64")
```

```python
employed = []
for i in df["days_employed"]:
    z = i / 365
    employed.append(z)
```

```python
df["years_employed"] = employed
```

```python
df["years_employed"] = df["years_employed"].abs().astype("int64")
```

```python
df.info()
```

```python
df["num_of_children"] = df["cnt_children"]
df["gender"] = df["code_gender"]
df["own_car"] = df["flag_own_car"]
df["own_real_estate"] = df["flag_own_realty"]
df["income_total"] = df["amt_income_total"]
df["income_type"] = df["name_income_type"]
df["education"] = df["name_education_type"]
df["family_status"] = df["name_family_status"]
df["housing_type"] = df["name_housing_type"]
df["num_of_family_members"] = df["cnt_fam_members"]
```

```python
del df["cnt_children"], df["code_gender"], df["flag_own_car"], df["flag_own_realty"], df["amt_income_total"], df["name_income_type"], df["name_education_type"], df["name_family_status"], df["name_housing_type"], df["cnt_fam_members"], df["days_birth"], df["days_employed"], df["flag_mobil"], df["flag_work_phone"], df["flag_phone"], df["flag_email"]  
```

```python
df["occupation_type"] = df["occupation_type"].fillna("Unknown")
```

```python
# may erase the unknowns at a later date - exceptionally high number - probably hurt the model
df["occupation_type"].value_counts()
```
