```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt

import tensorflow.keras
from tensorflow.keras.models import load_model
```

```python
knee_model = load_model("callback_model.h5")
```

```python
knee_dict = {0:'Healthy', 1:'Doubtful', 2:'Minimal', 3:'Moderate', 4:'Severe'}
```

```python
## First test image
img_size = 128
path = ("test_images/label_0_knee.png")
test_image = cv2.imread(os.path.join(path))[...,::-1]
resized_arr = cv2.resize(test_image, (img_size, img_size))
img = np.array(resized_arr) / 255
img = img.reshape(-1, img_size, img_size, 3)
knee_predict = knee_model.predict(img)[0]
knee_idx = np.argmax(knee_predict)
knee_cat = knee_dict[knee_idx]
plt.imshow(test_image)
print(knee_idx, knee_cat)
```
