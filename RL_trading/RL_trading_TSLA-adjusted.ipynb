```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import gym
import gym_anytrading
from gym_anytrading.envs import StocksEnv

from finta import TA

from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines import A2C
from stable_baselines.common.callbacks import CheckpointCallback, EvalCallback
```

```python
df = pd.read_csv('gdrive/My Drive/stock/TSLA.csv')
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.set_index('Date', inplace=True)
```

```python
df = df.sort_values(by=["Date"])
```

```python
df["Volume"] = df["Volume"].astype(float)
```

```python
df['SMA'] = TA.SMA(df, 10)
df['RSI'] = TA.RSI(df)
df['OBV'] = TA.OBV(df)
df.fillna(0, inplace=True)
```

```python
del df["Adj Close"]
```

```python
df
```

```python
## Random Action
```

```python
env = gym.make('stocks-v0', df=df, frame_bound=(290,300), window_size=5)
```

```python
env.action_space
```

```python
state = env.reset()
while True: 
    action = env.action_space.sample()
    n_state, reward, done, info = env.step(action)
    if done: 
        print("info", info)
        break
        
plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()

print("max_possible_profit:", env.max_possible_profit())
```

```python
# First model
```

```python
eval_callback = EvalCallback(env, best_model_save_path='gdrive/My Drive/stock/logs/',
                             log_path='gdrive/My Drive/stock/logs/', eval_freq=500,
                             deterministic=True, render=False)

checkpoint_callback = CheckpointCallback(save_freq=500, save_path='gdrive/My Drive/stock/logs/',
                                         name_prefix='rl_model')
```

```python
env_maker = lambda: gym.make('stocks-v0', df=df, frame_bound=(50,len(df)), window_size=5)
env = DummyVecEnv([env_maker])
```

```python
model = A2C('MlpLstmPolicy', env, verbose=1) 
model.learn(total_timesteps=100000, callback=eval_callback)
```

```python
env = gym.make('stocks-v0', df=df, frame_bound=(290,300), window_size=5)
obs = env.reset()
while True: 
    obs = obs[np.newaxis, ...]
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    if done:
        print("info", info)
        break

plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()

print("max_possible_profit:", env.max_possible_profit())
```
