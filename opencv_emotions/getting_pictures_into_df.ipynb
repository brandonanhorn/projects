```python
rootdir = "gdrive/My Drive/emotions_data"
```

```python
all_images = []
for subdir, dirs, files in os.walk(rootdir):
  for file in files:
    all_images.append(os.path.join(subdir,file))
```

```python
all_images
```

```python
df = pd.DataFrame()
```

```python
df["image"] = all_images
```

```python
df.head()
```

```python
df_1 = df.image.str.split("/",expand=True)
```

```python
df_1
```

```python
del df_1[0]
del df_1[1]
del df_1[2]
del df_1[4]
```

```python
df["emotion"] = df_1[3]
```

```python
df.head(1)
```

```python
df.tail(1)
```

```python
df = df.drop(df.index[0])
```

```python
df.head(1)
```

```python
df_images = []
for i in tqdm(df["image"]):
    img = image.load_img(i, target_size=(64,64,1), grayscale=False)
    img = image.img_to_array(img)
    img = img/255
    df_images.append(img)
```

```python
df["array"] = df_images
```
